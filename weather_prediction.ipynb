```python
def create_predictions(predictors, main, reg):
    train = main.loc[:"2020-12-13"]
    test = main.loc["2021-01-01":]
    reg.fit(train[predictors],train['target'])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"],predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)],axis=1)
    combined.columns = ["actual","predictions"]
    return error, combined
```

```python
main["month_max"] = main["temp_max"].rolling(30).mean()
```

```python
main
```

```python
main["month_day_max"] = main["month_max"] / main["temp_max"]
```

```python
main["max_min"] = main["temp_max"] / main["temp_min"]
```

```python
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
main = main.iloc[30:,:].copy()
```

```python
error, combined = create_predictions(predictors, main, reg)
```
